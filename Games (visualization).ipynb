```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
df=pd.read_csv('appstore_games.csv')
```

```python
df.head()
```

```python
df.drop('URL',axis=1,inplace=True)
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df.drop(['ID','Icon URL'],axis=1,inplace=True)
```

```python
sns.boxplot(y='Average User Rating',x='Age Rating',data=df,palette='rainbow')
```

```python
df1=df.dropna(axis=0)
```

```python
sns.barplot(x='Name',y='Price',data=df1)
```

```python
sns.barplot(x='Name',y='Price',data=df1,estimator=np.std)
```

```python
df1.head()
```

```python
sns.barplot(x='Average User Rating',y='User Rating Count',data=df1)
```

```python
#hence lower ratings recieve less user counts as not many users play games with lower ratings 
```
